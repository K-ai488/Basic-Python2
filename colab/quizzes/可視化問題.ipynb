```python
df = pd.read_csv(R'C:\Users\Tomoki Kubo\OneDrive\Desktop\playground\Basic-Python2\colab\quizzes\matplotlib_quiz\hist_data.csv') #前にRつけないと\がエスケープシーケンスとして見なされない
a,b = df['aデータ'], df['bデータ']

plt.hist(a, color = "orange", bins = 30, label = "aデータ", alpha = 0.4)
plt.hist(b, color = "darkturquoise", bins = 30, label = "bデータ", alpha = 0.4) #alphaを調整することで重なった部分の色を調節
plt.title('aデータとbデータの温度に関する度数分布')
plt.xlabel('温度（℃）')
plt.ylabel('度数')
plt.legend() #凡例
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

問2.png

```python
df = pd.read_csv(R'C:\Users\Tomoki Kubo\OneDrive\Desktop\playground\Basic-Python2\colab\quizzes\matplotlib_quiz\scatter_data.csv') #前にRつけないと\がエスケープシーケンスとして見なされない

class0,class1,class2 = df[df["class"] == 0],df[df["class"] == 1],df[df["class"] == 2]
plt.scatter(class0["length (cm)"],class0["width (cm)"],c="red",label="class0")
plt.scatter(class1["length (cm)"],class1["width (cm)"],c="blue",label="class1")
plt.scatter(class2["length (cm)"],class2["width (cm)"],c="green",label="class2")

plt.title("クラス0～2のlengthとwidthに関する散布図")
plt.xlabel("length (cm)")
plt.ylabel("width (cm)")

plt.legend()
plt.grid()#格子を入れる
plt.show()
```

```python
df = pd.read_csv(R'C:\Users\Tomoki Kubo\OneDrive\Desktop\playground\Basic-Python2\colab\quizzes\matplotlib_quiz\line_data.csv') #前にRつけないと\がエスケープシーケンスとして見なされない

period,number,cost = df['期間'], df['件数'], df['費用']

plt.title("期間ごとの件数の推移 ")
plt.xlabel("期間")
plt.ylabel("件数")

plt.plot(period,number,marker="o")
plt.grid()#格子を入れる
plt.show()
```
